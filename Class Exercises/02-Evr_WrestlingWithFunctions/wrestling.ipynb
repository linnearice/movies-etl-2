```python
import os
import csv
import pandas as pd

# Path to collect data from the Resources folder
data_to_load_df = os.path.join("resources/WWE-Data-2016.csv")
wrestling_data_df=pd.read_csv(data_to_load_df)
wrestling_data_df.head()
```

```python
# Define the function and have it accept the 'wrestlerData' as its sole parameter

# Find the total number of matches wrestled
total_matches = wrestling_data_df["Wins"] + wrestling_data_df["Losses"] + wrestling_data_df["Draws"]
wrestling_data_df["Total Matches"] = total_matches
wrestling_data_df.head()
```

```python
# Find the percentage of matches won
wins_percent=wrestling_data_df["Wins"]/wrestling_data_df["Total Matches"]*100
wrestling_data_df["Percent Wins"]= wins_percent
wrestling_data_df.head()
```

```python
# Find the percentage of matches lost
loss_percent=wrestling_data_df["Losses"]/wrestling_data_df["Total Matches"]*100
wrestling_data_df["Percent Losses"]= loss_percent
wrestling_data_df.head()
```

```python
# Find the percentage of matches drawn
drawn_percent=wrestling_data_df["Draws"]/wrestling_data_df["Total Matches"]*100
wrestling_data_df["Percent Drawn"]= drawn_percent
wrestling_data_df.head()
```
